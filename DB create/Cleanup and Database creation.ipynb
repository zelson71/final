```python
import pandas as pd
import glob
from ipywidgets.embed import embed_minimal_html

```

```python
path = 'Resources/WE_data'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, encoding = 'utf-8')
    li.append(df)

all_wines_df = pd.concat(li, axis=0, ignore_index=True,sort=True)
```

```python
all_wines_df.index.name = ('ID')
```

## Drop Wines with no Price

```python
all_wines_df = all_wines_df.dropna(subset=['price'],how='all')
```

## Drop Wines with no Country

```python
all_wines_df = all_wines_df.dropna(subset=['country'],how='all')
```

## Drop Wines with no alcohol information 

```python
all_wines_df = all_wines_df.dropna(subset=['alcohol'],how='all')
```

## Drop wines without Vintage and Varietal information

```python
all_wines_df = all_wines_df.dropna(subset=['varietal'],how='all')
all_wines_df = all_wines_df.dropna(subset=['vintage'],how='all')
```

## Copy data from subregion or subsubregion to the region if it is null

```python
all_wines_df.loc[all_wines_df['region'].isnull(),'region'] = all_wines_df['subregion']

all_wines_df.loc[all_wines_df['region'].isnull(),'region'] = all_wines_df['subsubregion']
```
