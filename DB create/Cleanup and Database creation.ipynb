```python
path = 'Resources/WE_data'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, encoding = 'utf-8')
    li.append(df)

all_wines_df = pd.concat(li, axis=0, ignore_index=True,sort=True)
```
